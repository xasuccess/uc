```python
for i in range(1,31):
    if i ==18:
        break       #终止函数
    if i%2==1:           # 百分号%取余数，打印偶数数列
        print("{0}  是 奇数".format(i))
    else:
        print("{0}  是 偶数".format(i))
```

```python
capital = 10000
year = 0
rest = 0.12
while capital<1000000:
    year +=1
    interrest = capital*rest
    capital = capital*(1+rest)
    print("第{0}年拿到了{1}元钱,利息是{2}元".format(year,capital,interrest))
```

```python
capital = 10000
year = 0
rest = 0.03
interrest = capital*rest
while interrest<10000:
    year +=1
    interrest = capital*rest
    capital = capital*(1+rest)
    print("第{0}年拿到了{1}元, 利息是{2}元".format(year,capital,interrest))
else:
    print("第{0}年增长利息超过本金".format(year))
```
