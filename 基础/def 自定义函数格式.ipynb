```python
#收集参数 ，把无法定义的参数放到一个函数 *args中
def stu(*args):
    print("hello，大家好，首先我来做个自我介绍：")
    for i in args:
        print(i, end = " ")

for a in range(1,2):
    stu(a)

    
```

```python
#收集参数 ，把无法定义的参数放到一个函数 *args中
def stu(*args):  #收集参数 约定 用  *args，也可以使用其他定义字符
    for i in args:
        print(i)   # 输出i，i数据来自列表
        print(args)   #输出 args 列表
        print(type(args))  #输出args格式

stu(1,23,4,35,4,6,66)
```

```python
#收集参数
def stu (**kwargs):
    for i,j in kwargs.items():
        print(i, "..", j)         #输出字典类型数据
        print(kwargs)            #打印kwargs
        print(type(kwargs))    #打印kwargs 数据类型 ，输出dict
        
stu(name = "dl", sex = "男")
stu()
    
```

```python
#混合参数，顺序 普通参数 关键字参数 收集参数tuple 收集参数dict
def stu(name = "dl", age = 18, hobby = "没有", *args, **kwargs):
    print("hello 大家好")
    print("我叫{0},我今年{1}岁。".format(name, age))
    if hobby == "没有":
        print("我没有爱好，哈哈哈")
    else:
        print("我的爱好是{0}".format(hobby))
    for i in args:
        print(i)
    for k,v in kwargs.items():
        print(k, "..", v)
        
stu("dl", 22, "打豆豆", "34","308", 补充哎dd = "df", 补充哎ddd = "df")
```

```python
#斐波那契数列
def fib(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fib(n-2) + fib(n-1)
    """ddeee """
for i in range(1,10):
    print(fib (i), end=" ")
```
